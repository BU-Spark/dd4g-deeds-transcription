```python
import pandas as pd

# Function to categorize document types
def categorize_document(doc_type):
    if pd.isna(doc_type):
        return "Unknown"
    
    doc_type = doc_type.lower()

    if "mortgage" in doc_type:
        return "Mortgages"
    
    elif "lien" in doc_type:
        return "Liens"
    
    elif "easement" in doc_type or "agreement" in doc_type:
        return "Easements"
    
    elif "plan" in doc_type or "map" in doc_type:
        return "Plan Cards"
    
    elif "deed" in doc_type or "conveyance" in doc_type or "sale" in doc_type or "partition" in doc_type or "gift" in doc_type or "property" in doc_type:
        return "Deeds"
    
    else:
        return "Other"


# Separate column for standardizing only Deeds (keeping "Document Type" intact)
df["Deeds Standardized"] = df["Document Type"].apply(lambda x: "Deeds" if pd.notna(x) and any(word in x.lower() for word in ["deed", "conveyance", "sale", "partition", "gift", "property"]) else None)

df.head(7)

```
