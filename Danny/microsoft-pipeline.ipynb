```python
import os
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image

# Specify the directory containing the images
image_folder = '/projectnb/sparkgrp/dd4g-reg-of-deeds/Danny/train-images1'

# List all files in the directory
image_files = os.listdir(image_folder)

# Choose the first image file (assuming it is an image file)
image_file = next((f for f in image_files if f.lower().endswith(('.png', '.jpg', '.jpeg'))), None)

# Check if an image file is found
if image_file:
    image_path = os.path.join(image_folder, image_file)  # Construct the full path to the image
    image = Image.open(image_path).convert("RGB")

    # Load the processor and model
    processor = TrOCRProcessor.from_pretrained('microsoft/trocr-base-handwritten')
    model = VisionEncoderDecoderModel.from_pretrained('microsoft/trocr-base-handwritten')

    # Process the image into pixel values
    pixel_values = processor(images=image, return_tensors="pt").pixel_values

    # Generate the text from the image
    generated_ids = model.generate(pixel_values)
    generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

    # Output the generated text
    print(generated_text)
else:
    print("No image files found in the directory.")

image.show
```
